```python
## Consumer

from kafka import KafkaConsumer
from pyspark.sql import SparkSession
from datetime import datetime

# Kafka server and topic information
bootstrap_servers = 'w01.itversity.com:9092'
topic = 'Ecomm_dataMng_Pipeline14'

# Create a Kafka consumer
consumer = KafkaConsumer(topic,bootstrap_servers = bootstrap_servers, group_id = 'Ecomm_datapipeline')

# Create empty lists to store the processed data
customer_data = []
product_data = []
transaction_data = []
inventory_data = []
promotion_data = []
store_location_data = []
supplier_data = []
delivery_data = []

# Global counter for data fetch
data_counter = 0

# consume and process the data
for message in consumer:
    # Get the consumed message value 
    message_value = message.value.decode('utf-8')
    #print(message_value)
    
    # Split the message value by comma (assuming comma-separated values)
    values = message_value.split(',')
    
    # Process the values and convert them to the desired data types
     ############################################ Indian customer DATA ####################################################### 
    customer_id = int(values[0])
    first_name = str(values[1])
    last_name = str(values[2])
    customer_email = str(values[3])
    customer_address = str(values[4])
    customer_phone_number = int(values[5])
    registration_date_str = str(values[6])
    registration_date = datetime.strptime(registration_date_str, '%Y-%m-%d').date()
    last_purchase_date_str = str(values[7])
    last_purchase_date = datetime.strptime(last_purchase_date_str, '%Y-%m-%d').date()
    gender = str(values[8])
    customer_age = int(values[9])
    date_of_birth_str = str(values[10])
    date_of_birth = datetime.strptime(date_of_birth_str, '%Y-%m-%d').date()
    #print(customer_id, first_name, last_name, customer_email, ',',customer_address,',', customer_phone_number,registration_date, last_purchase_date, gender, date_of_birth)
    nationality = str(values[11])
    occupation = str(values[12])
    preferred_language = str(values[13])
    #print(preferred_language)
    loyalty_points = int(values[14])
    registration_channel = str(values[15])
    #print(nationality, occupation, preferred_language, loyalty_points, registration_channel)
    marketing_opt_in = str(values[16])
    last_login_date_str = str(values[17])
    last_login_date = datetime.strptime(last_login_date_str, '%Y-%m-%d').date()
    subscription_status = str(values[18])
    loyalty_tier = str(values[19])
    #print(marketing_opt_in, last_login_date, subscription_status, loyalty_tier)
    #break
    ############################################ PRODUCT DATA #######################################################
    product_id = str(values[20])    
    product_name = str(values[21])
    product_main_category = str(values[22])
    product_sub_category = str(values[23])
    product_sub_sub_category = str(values[24])
    brand = str(values[25])
    description = str(values[26])
    price = float(values[27])
    discounted_price = float(values[28])
    discount_percentage = float(values[29])
    stock_quantity = int(values[30])
    created_date_str = str(values[31])
    created_date = datetime.strptime(created_date_str, '%Y-%m-%d').date()
    product_supplier_id = int(values[32])
    average_rating = float(values[33])
    customer_reviews_count = int(values[34])
    is_featured = str(values[35])
    is_new_arrival = str(values[36])
    is_on_sale = str(values[37])
    
    ######################################################## TRANSACTIONAL DATA ##########################################################
    transaction_id = str(values[38])
    transactional_customer_id = int(values[39])
    transactional_product_id = str(values[40]) 
    quantity = int(values[41])
    transactional_timestamp_str = str(values[42])
    transactional_timestamp = datetime.strptime(transactional_timestamp_str, '%Y-%m-%d').date()
    payment_method = str(values[43])
    subtotal = float(values[44])
    tax_amount = float(values[45])
    total_amount = float(values[46])
    transactional_shipping_address = str(values[47])
    billing_address = str(values[48])
    transactional_delivery_notes = str(values[49])
    transactional_is_gift = str(values[50])
    transactional_gift_message = str(values[51])
    is_returned = str(values[52])
    return_reason = str(values[53])
    refunded_amount = float(values[54])
    currency = str(values[55])
    exchange_rate = float(values[56])
    order_status = str(values[57])
    fulfillment_status = str(values[58])
    is_cancelled = str(values[59])
    cancellation_reason = str(values[60])
    payment_status = str(values[61])
    shipping_method = str(values[62])
    tax_rate = float(values[63])
    
    ###################################################### INVENTORY DATA #############################################################
    inventory_product_id = str(values[64])
    inventory_location = str(values[65])
    last_updated_inventory_str = str(values[66])
    last_updated_inventory = datetime.strptime(last_updated_inventory_str, '%Y-%m-%d').date()
    replenishment_threshold = int(values[67])
    maximum_stock_quantity = int(values[68])
    minimum_stock_quantity = int(values[69])
    reserved_quantity = int(values[70])
    damaged_quantity = int(values[71])
    obsolete_quantity = int(values[72])
    restocking_date_str = str(values[73])
    restocking_date = datetime.strptime(restocking_date_str, '%Y-%m-%d').date()
    stock_alerts_enabled = str(values[74])
    stock_alert_threshold = int(values[75])
    stock_alert_notification_emails = str(values[76])
    stock_location = str(values[77])
    storage_conditions = str(values[78])
    storage_temperature = int(values[79])
    storage_humidity = int(values[80])
    is_batched = str(values[81])
    batch_number = int(values[82])
    ############################################################ PROMOTIONS DATA ########################################################
    promotion_id = str(values[83]) 
    promotion_product_id = str(values[84])
    start_date_str = str(values[85])
    start_date = datetime.strptime(start_date_str, '%Y-%m-%d').date()
    end_date_str = str(values[86])
    end_date = datetime.strptime(end_date_str, '%Y-%m-%d').date()
    promotion_discount_percentage = int(values[87])
    promotion_description = str(values[88])
    promotion_name = str(values[89])
    promotion_type = str(values[90])
    target_customers = str(values[91])
    #print(start_date,end_date, promotion_discount_percentage,promotion_description,promotion_name,promotion_type,target_customers,)
    redemption_limit = int(values[92])
    #print(redemption_limit )
    discount_type = str(values[93])
    applicable_products = str(values[94])
    #applicable_products_str = applicable_products_str.strip('[]')
    #applicable_products = [int(p_id) for p_id in applicable_products_str.split(';') if p_id]
    applicable_categories = str(values[95])
    #applicable_categories = [category.strip() for category in applicable_categories_str.split(';') if category]
    promotion_terms = str(values[96])
    #print(target_customers, redemption_limit, discount_type,applicable_products, applicable_categories,promotion_terms)

    ###################################################### STORELOCATIONS DATA ###########################################################
    store_id = int(values[97])
    store_name = str(values[98])
    store_address = str(values[99])
    city = str(values[100])
    state = str(values[101])
    country = str(values[102])
    #print(store_id,store_name,store_address,city,state,country)
    postal_code = int(values[103])
    manager_name = str(values[104])
    store_phone_number = int(values[105])
    opening_time_str = str(values[106])
    # Convert the string to a datetime.time object
    opening_time = datetime.strptime(opening_time_str, '%H:%M:%S').time()
    closing_time_str = str(values[107])
    closing_time = datetime.strptime(closing_time_str, '%H:%M:%S').time()
    store_capacity = int(values[108])
    store_area = int(values[109])
    store_rating = float(values[110])
    parking_availability = str(values[111])
    wheelchair_accessible = str(values[112])
    fitting_rooms_count = int(values[113])
    cash_desks_count = int(values[114])
    store_services = str(values[115])
    store_features = str(values[116])
    #print(store_services,store_features)
    ###################################################### SUPPLIERS DATA #############################################################
    supplier_id = int(values[117])
    supplier_name = values[118]
    contact_person = values[119]
    Supplier_phone_number = values[120]
    supplier_email = values[121]
    supplier_address = values[122]
    contact_number = values[123]
    company_website = values[124]
    #print(supplier_id,supplier_name,contact_person,Supplier_phone_number,supplier_email,supplier_address,contact_number,company_website)
    tax_id = int(values[125])
    business_type = values[126]
    payment_terms = values[127]
    preferred_contact_time_str = str(values[128])
    preferred_contact_time = datetime.strptime(preferred_contact_time_str, '%H:%M:%S').time()
    supplier_notes = values[129]
    supplier_rating = float(values[130])
    supplier_reviews_count = int(values[131])
    
    ###################################################### DeliveryTracking DATA ##########################################################
    delivery_id = int(values[132])
    order_id = int(values[133])
    carrier = values[134]
    status = values[135]
    estimated_delivery_date_str = values[136]
    estimated_delivery_date = datetime.strptime(estimated_delivery_date_str, '%Y-%m-%d').date()
    actual_delivery_date_str = values[137]
    actual_delivery_date = datetime.strptime(actual_delivery_date_str, '%Y-%m-%d').date()
    shipping_address_delivery = values[138]
    pickup_address = values[139]
    delivery_notes_delivery = values[140]
    is_gift_delivery = values[141]
    gift_message_delivery = values[142]
    delivery_vehicle_type = values[143]
    delivery_driver_name = values[144]
    driver_contact_number = values[145]
    delivery_rating = float(values[146])
    print("delivery_rating : ",delivery_rating)
                         
    ##################################################### Store data in Dataframe ##################################                    
    # Append the processed data to the respective lists
    customer_data.append((customer_id, first_name, last_name, customer_email, customer_address, customer_phone_number, registration_date,
                          last_purchase_date, gender, customer_age, date_of_birth, nationality, occupation,
                          preferred_language, loyalty_points, registration_channel, marketing_opt_in,
                          last_login_date, subscription_status, loyalty_tier))

    product_data.append((product_id, product_name, product_main_category, product_sub_category, product_sub_sub_category,
                         brand, description, price, discounted_price, discount_percentage, stock_quantity, created_date,
                         product_supplier_id, average_rating, customer_reviews_count, is_featured, is_new_arrival, is_on_sale))

    transaction_data.append((transaction_id, transactional_customer_id, transactional_product_id, quantity, transactional_timestamp, payment_method, subtotal,
                             tax_amount, total_amount, transactional_shipping_address, billing_address, transactional_delivery_notes, transactional_is_gift,
                             transactional_gift_message, is_returned, return_reason, refunded_amount, currency, exchange_rate,
                             order_status, fulfillment_status, is_cancelled, cancellation_reason, payment_status,
                             shipping_method, tax_rate))

    inventory_data.append((inventory_product_id, inventory_location, last_updated_inventory, replenishment_threshold,
                           maximum_stock_quantity, minimum_stock_quantity, reserved_quantity, damaged_quantity,
                           obsolete_quantity, restocking_date, stock_alerts_enabled, stock_alert_threshold,
                           stock_alert_notification_emails, stock_location, storage_conditions, storage_temperature,
                           storage_humidity, is_batched, batch_number))

    promotion_data.append((promotion_id, promotion_product_id, start_date, end_date, promotion_discount_percentage,
                           promotion_description, promotion_name, promotion_type, target_customers, redemption_limit,
                           discount_type, applicable_products, applicable_categories,promotion_terms))

    store_location_data.append((store_id, store_name, store_address, city, state, country, postal_code, manager_name,
                                store_phone_number, opening_time, closing_time, store_capacity, store_area, store_rating,
                                parking_availability, wheelchair_accessible, fitting_rooms_count, cash_desks_count,
                                store_services, store_features))
    supplier_data.append((supplier_id, supplier_name, contact_person, Supplier_phone_number, supplier_email, supplier_address, 
                          contact_number, company_website, tax_id, business_type, payment_terms, preferred_contact_time, 
                          supplier_notes, supplier_rating, supplier_reviews_count))
    delivery_data.append((delivery_id, order_id, carrier, status, estimated_delivery_date,
                          actual_delivery_date_str, actual_delivery_date, shipping_address_delivery, pickup_address, 
                          delivery_notes_delivery, is_gift_delivery, gift_message_delivery, delivery_vehicle_type, 
                          delivery_driver_name, driver_contact_number, delivery_rating))
    
    # Break the loop if enough data has been processed (e.g., 1000 rows)
    if data_counter >= 1000:
        break

# Close the Kafka consumer
consumer.close()

# Create a SparkSession
spark = SparkSession.builder \
    .appName("Ecomm data Management") \
    .getOrCreate()

# Create DataFrames from the lists of processed data
customer_df = spark.createDataFrame(
    customer_data,
    ["customer_id", "first_name", "last_name", "email", "address", "phone_number", "registration_date",
     "last_purchase_date", "gender", "Age", "date_of_birth", "nationality", "occupation", "preferred_language",
     "loyalty_points", "registration_channel", "marketing_opt_in", "last_login_date", "subscription_status",
     "loyalty_tier",]
)  


product_df = spark.createDataFrame(
    product_data,
    ["product_id", "product_name", "product_main_category", "product_sub_category", "product_sub_sub_category",
     "brand", "description", "price", "discounted_price", "discount_percentage", "stock_quantity", "created_date",
     "supplier_id", "average_rating", "customer_reviews_count", "is_featured", "is_new_arrival", "is_on_sale"]
)

transaction_df = spark.createDataFrame(
    transaction_data,
    ["transaction_id", "customer_id", "product_id", "quantity", "timestamp", "payment_method", "subtotal",
     "tax_amount", "total_amount", "shipping_address", "billing_address", "delivery_notes", "is_gift", "gift_message",
     "is_returned", "return_reason", "refunded_amount", "currency", "exchange_rate", "order_status", "fulfillment_status",
     "is_cancelled", "cancellation_reason", "payment_status", "shipping_method", "tax_rate"]
)

inventory_df = spark.createDataFrame(
    inventory_data,
    ["product_id", "location", "last_updated_inventory", "replenishment_threshold", "maximum_stock_quantity",
     "minimum_stock_quantity", "reserved_quantity", "damaged_quantity", "obsolete_quantity", "restocking_date",
     "stock_alerts_enabled", "stock_alert_threshold", "stock_alert_notification_emails", "stock_location",
     "storage_conditions", "storage_temperature", "storage_humidity", "is_batched", "batch_number"]
)

promotion_df = spark.createDataFrame(
    promotion_data,
    ["promotion_id", "product_id", "start_date", "end_date", "promotion_discount_percentage", "promotion_description",
     "promotion_name", "promotion_type", "target_customers", "redemption_limit", "discount_type", "applicable_products",
     "applicable_categories"]
)

store_location_df = spark.createDataFrame(
    store_location_data,
    ["store_id", "store_name", "store_address", "city", "state", "country", "postal_code", "manager_name",
     "store_phone_number", "opening_time", "closing_time", "store_capacity", "store_area", "store_rating",
     "parking_availability", "wheelchair_accessible", "fitting_rooms_count", "cash_desks_count", "store_services",
     "store_features"]
)
supplier_df = spark.createDataFrame(
    supplier_data,
    ["supplier_id", "supplier_name", "contact_person", "phone_number_supplier", "email_supplier", "supplier_address", 
     "contact_number", "company_website", "tax_id", "business_type", "payment_terms", "preferred_contact_time", 
     "supplier_notes", "supplier_rating", "supplier_reviews_count"]
)
delivery_df = spark.createDataFrame(
    delivery_data,
    ["delivery_id", "order_id", "carrier, status", "estimated_delivery_date_str", "estimated_delivery_date",
     "actual_delivery_date_str", "actual_delivery_date", "shipping_address_delivery", "pickup_address", "delivery_notes_delivery",
     "is_gift_delivery", "gift_message_delivery", "delivery_vehicle_type", "delivery_driver_name", "driver_contact_number", "delivery_rating"]
)                         

# Show the DataFrames
customer_df.show(truncate=False)
product_df.show(truncate=False)
transaction_df.show(truncate=False)
inventory_df.show(truncate=False)
promotion_df.show(truncate=False)
store_location_df.show(truncate=False)
supplier_df.show(truncate=False)
delivery_df.show(truncate=False)

# Define the HDFS output path
hdfs_output_path = '/user/itv007039/warehouse/ybm_test_db.db/'

# List of DataFrames
dataframes = [customer_df, product_df, transaction_df, inventory_df, promotion_df, store_location_df, supplier_df, delivery_df]

# Save DataFrames to HDFS
for df in dataframes:
    # Get DataFrame name
    df_name = df.schema.fieldNames()[0]  # Assuming the first column is the identifier
    # Write the DataFrame to HDFS
    df.write.parquet(hdfs_output_path + f'{df_name}_data.parquet', mode='overwrite')
                
# Stop the SparkSession
spark.stop()                     
```
